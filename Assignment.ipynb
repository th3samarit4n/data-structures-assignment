# QUESTION TWO: Algorithm Analysis and Complexity

## (a) Define the following asymptotic notations

### i. Big-O Notation (O)
Big-O notation describes the **worst-case time complexity** of an algorithm. It shows the maximum time an algorithm can take as input size increases.

### ii. Big-Theta Notation (Θ)
Big-Theta notation describes the **exact growth rate** of an algorithm. It gives both upper and lower bounds.

### iii. Big-Omega Notation (Ω)
Big-Omega notation describes the **best-case time complexity** of an algorithm. It shows the minimum time required for an algorithm to execute.

## (b) Analyze the following code segment

```python
for i in range(n):
    for j in range(n):
        print(i, j)
```

### i. Time Complexity Analysis

The outer loop runs **n times** and the inner loop also runs **n times**.

Therefore:

**n × n = n²**

Hence, the time complexity is **O(n²)**.

### ii. Is the algorithm efficient for large values of n?

No, the algorithm is not efficient for large values of `n` because execution time increases quadratically. As `n` becomes larger, the number of operations increases significantly, making the program slower.

# QUESTION THREE: Searching and Sorting Algorithms

Given List:

```python
arr = [21, 10, 45, 32, 67, 5, 18]
```

## (a) Linear Search to find 32

Linear search checks each element one by one until the target is found.

````python
# Linear search function
def linear_search(arr, target):
    for index in range(len(arr)):
        if arr[index] == target:
            return index
    return -1


arr = [21, 10, 45, 32, 67, 5, 18]

result = linear_search(arr, 32)

if result != -1:
    print("Value found at index:", result)
else:
    print("Value not found")
````

## (b) Binary Search to find 45

Binary search works by repeatedly dividing a sorted list into halves.

Pseudocode:

```text
START
SORT the list
SET low = 0
SET high = length of list - 1

WHILE low <= high
    FIND middle element

    IF middle element equals target
        RETURN index
    ELSE IF target is larger
        SEARCH right half
    ELSE
        SEARCH left half
END
```

````python
# Binary search function
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1


arr = [21, 10, 45, 32, 67, 5, 18]

arr.sort()

print("Sorted List:", arr)

result = binary_search(arr, 45)

if result != -1:
    print("Value found at index:", result)
else:
    print("Value not found")
````

## (c) Bubble Sort in Ascending Order

Bubble sort repeatedly compares adjacent elements and swaps them if they are in the wrong order.

````python
# Bubble sort function
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr


arr = [21, 10, 45, 32, 67, 5, 18]

sorted_arr = bubble_sort(arr)

print("Sorted List:", sorted_arr)
````

# QUESTION FOUR: Linked Lists Implementation

## (a) Define a Node class for a singly linked list

A node in a singly linked list contains:

- **Data** → stores the value
- **Next pointer** → points to the next node

````python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
````

## (b) LinkedList class operations

The linked list should support:

1. Insert at the beginning  
2. Insert at the end  
3. Display all elements

### Pseudocode

```text
START

CREATE Node class

CREATE LinkedList class

INSERT at beginning:
    CREATE new node
    POINT new node to head
    UPDATE head

INSERT at end:
    MOVE to last node
    ADD new node

DISPLAY all elements

END
```

````python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Linked List class
class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Insert at end
    def insert_end(self, data):
        new_node = Node(data)

        # If list is empty
        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_node

    # Display linked list
    def display(self):
        current = self.head

        while current:
            print(current.data, end=" -> ")
            current = current.next

        print("None")


# Create linked list
ll = LinkedList()

# Insert elements
ll.insert_beginning(10)
ll.insert_beginning(5)

ll.insert_end(20)
ll.insert_end(30)

# Display list
print("Linked List:")
ll.display()
````

# QUESTION FIVE: Application of Data Structures

### Scenario
A university wants to develop a system that manages students waiting to be served at the finance office.

## (a) Most appropriate data structure and justification

The most appropriate data structure is a **Queue**.

### Justification

A queue follows the **FIFO (First In First Out)** principle.

This means:

- The first student to arrive is the first student served.
- New students join at the back of the queue.

This makes a queue ideal for managing waiting lines in a finance office.

## (b) Python Program to Simulate the Queue System

### Pseudocode

```text
START

CREATE queue

ADD student to queue

REMOVE first student from queue

DISPLAY queue

END
```

````python
from collections import deque


# Queue class
class StudentQueue:
    def __init__(self):
        self.queue = deque()

    # Add student
    def add_student(self, student):
        self.queue.append(student)
        print(f"{student} added to the queue.")

    # Serve student
    def serve_student(self):
        if len(self.queue) == 0:
            print("No students in queue.")
        else:
            student = self.queue.popleft()
            print(f"{student} has been served.")

    # Display queue
    def display_queue(self):
        if len(self.queue) == 0:
            print("Queue is empty.")
        else:
            print("Current Queue:")
            for student in self.queue:
                print(student)


# Create queue
finance_queue = StudentQueue()

# Add students
finance_queue.add_student("John")
finance_queue.add_student("Mary")
finance_queue.add_student("David")

# Display queue
finance_queue.display_queue()

# Serve student
finance_queue.serve_student()

# Display updated queue
finance_queue.display_queue()
````
